```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
```

```python
data= pd.read_csv('/content/diabetes.csv')
data.head()
```

```python
# Plot correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Distribution of target variable
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x=data.columns[-1], palette='viridis')
plt.title('Target Variable Distribution')
plt.xlabel('Outcome (0 = Non-Diabetic, 1 = Diabetic)')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of individual features
features = data.columns[:-1]
plt.figure(figsize=(15, 12))
for i, feature in enumerate(features):
    plt.subplot(3, 3, i + 1)
    sns.histplot(data[feature], kde=True, color='blue', bins=30)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# ----------------------- Modeling and Evaluation -----------------------

# Separate features and target
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
# Define models
dt_model = DecisionTreeClassifier(random_state=2)
svm_model = SVC(kernel='linear', random_state=2)
```

```python
# Define K values for Stratified K-Fold
k_values = [3, 4, 5, 6, 7, 8, 9, 10]
```

```python
# Metrics storage for both models
results = {
    'k': [],
    'Model': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1 Score': []
}
```

```python
# Function to evaluate a model with cross-validation for a given k
def evaluate_model_k(model, model_name, k):
    skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=2)
    accuracy = cross_val_score(model, X, y, cv=skf, scoring=make_scorer(accuracy_score)).mean()
    precision = cross_val_score(model, X, y, cv=skf, scoring=make_scorer(precision_score)).mean()
    recall = cross_val_score(model, X, y, cv=skf, scoring=make_scorer(recall_score)).mean()
    f1 = cross_val_score(model, X, y, cv=skf, scoring=make_scorer(f1_score)).mean()
```

```python
# Store results
results = {
    'k': [],
    'Model': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1 Score': []
}

```

```python
# Iterate through k values for both models
for k in k_values:
    evaluate_model_k(dt_model, 'Decision Tree', k)
    evaluate_model_k(svm_model, 'SVM', k)

# Convert results to DataFrame for visualization
results_df = pd.DataFrame(results)
```

```python
# Print the summary of metrics
print("\nSummary of Performance Metrics:\n")
print(results_df)

```

```python
# ----------------------- Visualization -----------------------

# Plot performance metrics as box plots
plt.figure(figsize=(14, 8))
sns.set(style='whitegrid')
```

```python
# Boxplot for each metric
for metric in ['Accuracy', 'Precision', 'Recall', 'F1 Score']:
    sns.boxplot(data=results_df, x='Model', y=metric, palette='Set2')
    plt.title(f'{metric} Comparison Between Models Across Different K Values')
    plt.xlabel('Model')
    plt.ylabel(metric)
    plt.show()
```

```python
# Plot metric trends as histograms
plt.figure(figsize=(14, 8))
for metric in ['Accuracy', 'Precision', 'Recall', 'F1 Score']:
    sns.histplot(data=results_df, x=metric, hue='Model', kde=True, palette='Set1', bins=15, alpha=0.7)
    plt.title(f'{metric} Distribution Across Different K Values')
    plt.xlabel(metric)
    plt.ylabel('Frequency')
    plt.show()
```
